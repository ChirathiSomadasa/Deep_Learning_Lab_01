3. Calculate the mean, standard deviation and quartile values for each of the above 4 features.

```python
# Select only the 4 required features
selected_features = ['N_total', 'N_bulk', 'N_surface', 'R_avg']
feature_df = df[selected_features]

# Calculate mean
print("Mean values:\n", feature_df.mean())

# Calculate standard deviation
print("\nStandard Deviation:\n", feature_df.std())

# Calculate quartile values (25%, 50%, 75%)
print("\nQuartiles:\n", feature_df.quantile([0.25, 0.5, 0.75]))
```

4. Plot the histogram of each of these features in a 1x4 layout.

```python
import matplotlib.pyplot as plt

features = ['N_total', 'N_bulk', 'N_surface', 'R_avg']

fig, axes = plt.subplots(1, 4, figsize=(20, 5))

# Use a regular for loop with index
for i in range(len(features)):
    feature = features[i]
    axes[i].hist(df[feature], bins=30, color='skyblue', edgecolor='black')
    axes[i].set_title(feature)
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

5. Visualize the scatter plots and histograms of this dataframe using the pairplot functionality
of seaborn library.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a dataframe with only the selected 4 features
plot_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]

# Generate pairplot (scatter plots + histograms)
sns.pairplot(plot_df)
plt.suptitle("Pairplot of N_total, N_bulk, N_surface, R_avg", y=1.02)
plt.show()
```

6. Add the below code and change it such that,

   a. Plots on the diagonal contains the histogram of each feature along with the kernal density estimation plot.

  b. Plots on the lower half [g.map_lower] contains the bivariate kernal density estimation plot.

  #new_df is the dataframe containing only the above mentioned 4 features.
  
  g = sns.PairGrid(new_df)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a new DataFrame with the 4 required features
new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]

# Initialize the PairGrid
g = sns.PairGrid(new_df)

# Upper triangle â€“ use bivariate histograms
g.map_upper(sns.histplot)

# Diagonal â€“ histogram + KDE
g.map_diag(sns.histplot, kde=True)

# Lower triangle â€“ bivariate KDE plots
g.map_lower(sns.kdeplot, fill=True)

# Show the plot
plt.suptitle("Customized PairGrid with KDE", y=1.02)
plt.show()
```

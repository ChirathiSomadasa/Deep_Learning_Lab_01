2. Create a random 100000x1 array from the same exponential distribution and visualize the distribution histogram along with uniform and normal distributions. Use the code given below and do the necessary changes. Change the number of bins and the plot view to make the visualization better

```python
import numpy as np
import matplotlib.pyplot as plt

n = 100000
scale = 1.0

exp_data = np.random.exponential(scale=scale, size=n)
uniform_data = np.random.rand(n)
normal_data = np.random.randn(n)

plt.hist(uniform_data, density=True, bins=200, histtype='step', color='blue', label='Uniform')
plt.hist(normal_data, density=True, bins=200, histtype='step', color='red', label='Normal')
plt.hist(exp_data, density=True, bins=200, histtype='step', color='green', label='Exponential')

plt.axis([0, 10, 0, 1.1])  # Adjust axis to show exponential tail properly
plt.legend(loc='upper right')
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

3. Using matplotlib, plot the Z = X**2 + Y**2 in a 3D plot. Limit the x and y values to [-5, 5].

```python
import matplotlib
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

figure = plt.figure(1, figsize = (12, 4))
subplot3d = plt.subplot(111, projection='3d')
surface = subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0.1)
plt.show()
```

4. Calculate the pearson standard coefficient and spearman rank correlation for the [HP, Attack, Defense, Sp. Ark, Sp. Def, Speed] features in the seaborn tutorial.ipynb. Visualize the results using heatmaps. Make sure the values are displaced in the heatmap as well.

```python
# Import libraries
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# Load the dataset
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
```

```python
# Select relevant numerical features
stats_df = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]
```

```python
# Pearson Correlation
pearson_corr = stats_df.corr(method='pearson')
print("Pearson Correlation Matrix:")
print(pearson_corr)
```

```python
# Spearman Correlation
spearman_corr = stats_df.corr(method='spearman')
print("\nSpearman Correlation Matrix:")
print(spearman_corr)
```

```python
# Plot Pearson heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(pearson_corr)
plt.title('Pearson Correlation Heatmap')
plt.show()
```

```python
# Plot Spearman heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(spearman_corr)
plt.title('Spearman Correlation Heatmap')
plt.show()
```
